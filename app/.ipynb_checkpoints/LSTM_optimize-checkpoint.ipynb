# 1 day ahead

```python
# run for adam optimizer only as it works better in this case than SGD
# run for a stock with long history, e.g. IBM
for n_steps in (60, 80):
    for epochs in (50, 70):
        for lr in (0.0005, 0.001):
            for l_size in (40, 60):
                    for test_size in (0.15, 0.2):
                            #try:
                            lstm = LstmModel()
                            lstm.get_input_data('IBM', offset=1)
                            lstm.normalize_data()
                            lstm.train_test_split(val_size = 0.4, test_size = test_size, n_steps=n_steps)
                            lstm.lstm_model(lr = lr, optimizer = 'adam', dropout=0.05, \
                                            l1_size=l_size, l2_size=l_size, l3_size=l_size)
                            lstm.fit_model(epochs=epochs, batch_size=64)
                            # predict
                            test_predict = lstm.lstm.predict(lstm.X_test)
                            test_predict = lstm.scaler.inverse_transform(test_predict)
                            # error
                            MSE = np.sqrt(mse(lstm.y_test, test_predict))
                            print('n_steps : {}, epochs: {}, lr: {}, l_size: {}, test_size: {}, MSE: {}'\
                                .format(n_steps, epochs, lr, l_size, test_size, MSE))
                            #except:
                            #    print('Couldnt run for n_steps : {}, epochs: {}, lr: {}, l_size: {}, opt: {}, test_size: {}, val_size: {}'\
                            #        .format(n_steps, epochs, lr, l_size, opt, test_size, val_size))
```
